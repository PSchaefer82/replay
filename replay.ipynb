```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://baseballsavant.mlb.com/replay?challenge_type=&year=2024&challenger=teamchallenging&team="
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
table = soup.find('table', class_='table')
tables = table.text
print(tables)
```

```python
tables
```

```python
# Create an empty list
# Loop through the scraped data to create a list of rows
table_facts = []
table_rows = table.find_all('tr')

for row in table_rows[1:]:
    row_data = row.find_all('td')
    table_facts.append({
        "replay_number" : row_data[row].text,
        "game_date": row_data[row].text,
        "game":row_data[row].text,
        "challenging_team":row_data[3].text,
        "challenge_type":row_data[4].text,
        "overturned?":row_data[5].text,
        "pitcher":row_data[6].text,
        "batter":row_data[7].text,
        "inning":row_data[8].text
    })
    print(table_facts)
```

```python
table_facts
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(table_facts)
```

```python
# Confirm DataFrame was created successfully
df
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

```python
# 1. How many months are there on Mars?
team_challenging = df["challenging_team"].nunique()
team_challenging
```

```python
# 2. How many sols (Martian days) worth of data are there?
mars_days = df["sol"].nunique()
mars_days
```

```python
# 3. What is the average minimum temperature by month?
month_min_temp = df.groupby('month')['min_temp'].mean()
print(month_min_temp)
```

```python
# Plot the average minimum temperature by month
plt.figure(figsize=(10, 6))
plt.bar(month_min_temp.index, month_min_temp.values, color='blue')
plt.title('Average Minimum Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Minimum Temperature (°F)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
sorted_avg_min_temp = month_min_temp.sort_values()
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
plt.figure(figsize=(8, 6))
plt.bar(month_min_temp.index, sorted_avg_min_temp.values, color='blue')
plt.title('Average Minimum Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Minimum Temperature (°F)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 4. What is the average pressure by month?
pressure_avg = df.groupby('month')['pressure'].mean()
pressure_avg

```

```python
# Plot the average pressure by month
plt.figure(figsize=(8, 6))
plt.bar(pressure_avg.index, pressure_avg, color='blue')
plt.title('Average Pressure by Month')
plt.xlabel('Month')
plt.ylabel('Average Pressure')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
sorted_pressure = pressure_avg.sort_values()
sorted_pressure
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
sorted_pressure = pressure_avg.sort_values()

plt.figure(figsize=(8, 6))
plt.bar(pressure_avg.index, sorted_pressure, color='blue')
plt.title('Average Pressure by Month')
plt.xlabel('Month')
plt.ylabel('Average Pressure')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
df.plot(x='terrestrial_date', y='min_temp', rot=90)
plt.ylabel('Minimum Temperature (°C)')
plt.xlabel('Terrestrial Date')
plt.grid()
```

#### Minimum Temperature

Average temperature is warmer in the second half of the year, with high points in the 8th and 9th months. Coldest tempertures occured in the third and fourth month. There is a span of 5 months between the coldest (3) and warmest (8) months for average minimum temperature.

All values seem to fall between -65 and -85. Months 8 and 9 were the only to have a temperature above -70, months 3 and 4 were the only to have temperatures below -80.

#### Atmospheric Pressure

Lowest atmospheric temperature readings were taken in the 5th and 6th months, highest were found in month 9. There is a second dip in values for the 1st and 12th month, but not as low as found in the 5th and 6th.  

All values seem to be within the 750 and 850 range with months 5-6-7 being the only to fall below 800.

#### Temp Patterns

Lows have values in the -85 to -90 range, and look to occure every other year, with high minimum temperatures of -65 to -70 occuring in the years between.

#### How many terrestrial days exist in a Martian year?

668 sols (Martian Solar Days) equalling 687 days on earth. We can see this from the graph of daily minimum temp where it alternates every two years, taking two years to return to a peak on the line graph.

For earth, the minimum temp would fall at the same time the next year.
